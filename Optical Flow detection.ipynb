```python
import cv2

# Global variables
ref_point = []
roi_selected = False

# Mouse callback function
def select_roi(event, x, y, flags, param):
    global ref_point, roi_selected

    if event == cv2.EVENT_LBUTTONDOWN:
        ref_point = [(x, y)]
        roi_selected = False

    elif event == cv2.EVENT_LBUTTONUP:
        ref_point.append((x, y))
        roi_selected = True

# Load video
video = cv2.VideoCapture("hello.mp4")

# Create a window and set mouse callback
cv2.namedWindow("Select ROI")
cv2.setMouseCallback("Select ROI", select_roi)

# Process the video frames
while True:
    # Read the current frame
    ret, frame = video.read()

    if not ret:
        break

    # Clone the frame to display the selected ROI
    display_frame = frame.copy()

    # Draw the selected ROI rectangle on the display frame
    if len(ref_point) == 2 and roi_selected:
        cv2.rectangle(display_frame, ref_point[0], ref_point[1], (0, 255, 0), 2)

        # Extract the ROI coordinates and dimensions
        x = min(ref_point[0][0], ref_point[1][0])
        y = min(ref_point[0][1], ref_point[1][1])
        width = abs(ref_point[1][0] - ref_point[0][0])
        height = abs(ref_point[1][1] - ref_point[0][1])

        # Display the ROI coordinates and dimensions on the frame
        cv2.putText(display_frame, f"ROI: ({x}, {y}), Width: {width}, Height: {height}", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Display the frame with the selected ROI
    cv2.imshow("Select ROI", display_frame)

    if cv2.waitKey(1) == ord('q'):  # Press 'q' to quit
        break

# Release the video capture and close the output window
video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load video
video = cv2.VideoCapture("hello.mp4")

# Read the first frame
ret, prev_frame = video.read()

# Select a region of interest (ROI) the object
x, y, width, height = 100, 100, 440,501 

# Create a mask for drawing optical flow vectors
mask = np.zeros_like(prev_frame)

# Set parameters for Lucas-Kanade optical flow
lk_params = dict(winSize=(15, 15), maxLevel=2, criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

# Define color for drawing optical flow vectors
color = (0, 255, 0)

while True:
    # Read the current frame
    ret, frame = video.read()
    
    if not ret:
        break
    
    # Convert frames to grayscale
    prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Select the ROI in the previous frame
    prev_roi = prev_gray[y:y+height, x:x+width]
    
    # Calculate optical flow using Lucas-Kanade method
    next_points, status, _ = cv2.calcOpticalFlowPyrLK(prev_roi, gray[y:y+height, x:x+width], None, None, **lk_params)
    
    # Select good points
    good_points = next_points[status == 1]
    
    # Draw optical flow vectors on the frame
    for point in good_points:
        x_pos, y_pos = point.ravel()
        cv2.circle(frame, (x + int(x_pos), y + int(y_pos)), 5, color, -1)
    
    # Draw rectangle around the object
    cv2.rectangle(frame, (x, y), (x + width, y + height), color, 2)
    
    # Update previous frame
    prev_frame = frame.copy()
    
    # Display the frame with optical flow
    cv2.imshow("Optical Flow", frame)
    
    if cv2.waitKey(1) == ord('q'):  # Press 'q' to quit
        break

# Release the video capture and close the output window
video.release()
cv2.destroyAllWindows()
```
